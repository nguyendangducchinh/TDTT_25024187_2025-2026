Xây dựng lớp Date biểu diễn ngày tháng năm với hai phương
thức kiểm tra xem một ngày tháng nào đó có hợp lệ hay không
và phương thức tính ngày kế tiếp của ngày tháng năm nào đó
nếu nó là hợp lệ
Nhập vào một ngày tháng năm, (ngày tháng năm biểu diễn
dưới dạng dd/mm/yyyy). Nếu ngày tháng năm không tồn tại, in
ra dòng chữ INVALID. Nếu là hợp lệ in ra ngày kế tiếp của nó.
Đầu vào
Đầu vào từ bàn phim chứa dữ liệu ngày, tháng, năm dưới dạng
dd/mm/yyyy
.
Đầu ra
In ra màn hình ngày kế tiếp của ngày vừa nhập dưới dạng
dd/mm/yyyy
. Nếu ngày tháng không tồn tại, in ra INVALID

```python
class MyDate :
    def __init__(self, _d, _m, _y):
        self.d = _d
        self.m = _m
        self.y = _y
    def checkValid(self):
        maxday = -1
        match self.m:
            case 1|3|5|7|8|10|12 :
                maxday = 31
            case 4|6|9|11 :
                maxday = 30
            case 2 :
                if (self.y % 4 == 0 and self.y % 100 != 0 ) or (self.y % 400 == 0 ):
                    maxday = 29
                else :
                    maxday = 28
        return (self.y > 0 and 1 <= self.d <= maxday)
    def getNextDay(self) -> MyDate:
        nextDay = MyDate(self.d + 1,self.m,self.y)
        if (nextDay.checkValid()):
            return nextDay 
        nextDay.d = 1
        nextDay.m += 1
        if (nextDay.checkValid()):
            return nextDay 
        nextDay.m = 1
        nextDay.y += 1
        return nextDay 
    def __str__(self):
        return f"{self.d:02d} / {self.m:02d} / {self.y}"

day, month, year = map(int,input().split('/'))
day1 = MyDate(day,month,year)
print(day1.checkValid())      
nextday = day1.getNextDay()
print(nextday)
```
