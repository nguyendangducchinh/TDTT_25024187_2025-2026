Cho một mảng số nguyên không âm. Gọi mảng đó là "mảng đẹp" nếu mỗi phần tử x trong mảng xuất hiện đúng x lần. Kiểm tra xem mảng có phải là mảng đẹp hay không. Nếu không hãy xóa đi y số (y là nhỏ nhất có thể) để mảng còn lại thành mảng đẹp.

```python
arr = list(map(int, input().split()))

counts = {}
for x in arr:
    counts[x] = counts.get(x, 0) + 1
removes = 0
for x, count in counts.items():
    if x == 0:
        removes += count
    elif count >= x:
        removes += count - x
    else:
        removes += count
print(removes)
```
