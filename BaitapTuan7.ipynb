### W7A3
Viết hàm thực hiện thuật toán bubble sort, Selection Sort để sắp xếp danh sách theo thứ tự tăng dần. Sắp xếp nổi bọt (Bubble Sort); Sắp xếp chọn (Selection Sort)

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr
arr = list(map(int, input().split()))
print("Bubble Sort:", bubble_sort(arr.copy()))
print("Selection Sort:", selection_sort(arr.copy()))
```

```python
def find_most_frequent(arr):
    if not arr:
        return None
    counts = {}
    for x in arr:
        counts[x] = counts.get(x, 0) + 1    
    max_item = arr[0]
    max_count = counts[arr[0]]
    
    for x in arr:
        if counts[x] > max_count:
            max_count = counts[x]
            max_item = x          
    return max_item
arr = list(map(int, input().split()))
print(find_most_frequent(arr))
```

### W7A5
Cho danh sách các số nguyên và một số X, đếm tất cả các cặp (a, b) sao cho a + b = X. c

```python
def count_pairs_with_sum(arr, X):
    seen = {}
    count = 0
    for num in arr:
        target = X - num
        if target in seen:
            count += seen[target]
        seen[num] = seen.get(num, 0) + 1
    return count
arr = list(map(int, input().split()))
X = int(input())
print(count_pairs_with_sum(arr, X))
```

```python
def flatten_list(nested_list):
    flat_list = []
    for sublist in nested_list:
        for item in sublist:
            flat_list.append(item)
    return flat_list

nested_list = [[1, 2], [3, 4], [5]]
print(flatten_list(nested_list))
```

### W7A7
Cho một danh sách các số nguyên nums, hãy tìm độ dài của dãy con tăng nghiêm ngặt dài nhất (không nhất thiết phải liền kề).

```python
def length_of_lis(nums):
    if not nums:
        return 0
    
    dp = [1] * len(nums)
    for i in range(len(nums)):
        for j in range(i):
            if nums[i] > nums[j]:
                dp[i] = max(dp[i], dp[j] + 1)
                
    return max(dp)

arr = list(map(int, input().split()))
print(length_of_lis(arr))
```

### W7A9
Cho một danh sách các số nums, với mỗi phần tử nums[i], hãy đếm xem có bao nhiêu phần tử bên phải nhỏ hơn nums[i]. Trả về danh sách các số đếm. Efficient solution: using heap or binary search tree

```python
import bisect

def countSmaller(nums):
    result = []
    sorted_list = []
    
    for num in reversed(nums):
        index = bisect.bisect_left(sorted_list, num)
        result.append(index)
        sorted_list.insert(index, num)
        
    return result[::-1]
arr = list(map(int, input().split()))
print(countSmaller(arr))
```

Cho một danh sách các số nguyên nums và một số nguyên m, hãy tìm tổng mảng con lớn nhất theo modulo m. Tức là, với bất kỳ mảng con nums[i:j] nào, hãy tính sum(nums[i:j]) % m và trả về kết quả lớn nhất có thể. Gợi ý: Duy trì tổng tiền tố (prefix sum) mod m và sử dụng tìm kiếm nhị phân để tìm tổng tiền tố nhỏ nhất chỉ lớn hơn tổng hiện tại một chút (để tối đa hóa hiệu số bao quanh).

```python
import bisect

def maxSubarrayMod(nums, m):
    max_val = 0
    prefix_sum = 0
    prefix_set = []
    
    for num in nums:
        prefix_sum = (prefix_sum + num) % m
        max_val = max(max_val, prefix_sum)
        
        idx = bisect.bisect_right(prefix_set, prefix_sum)
        if idx < len(prefix_set):
            max_val = max(max_val, (prefix_sum - prefix_set[idx] + m) % m)
            
        bisect.insort(prefix_set, prefix_sum)
        
    return max_val
arr = list(map(int, input().split()))
m = int(input())
print(maxSubarrayMod(arr, m))
```

```python
arr = list(map(int, input().split()))

counts = {}
for num in arr:
    counts[num] = counts.get(num, 0) + 1

max_length = 0
for num in counts:
    if num + 1 in counts:
        max_length = max(max_length, counts[num] + counts[num + 1])

print(max_length)
```

Đầu vào: số nguyên n - số lượng khách mời trong danh sách trên một dòng, n dòng tiếp theo là tên của các vị khách mời.
