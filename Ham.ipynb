### W5A3
Viết hàm kiểm tra một số có phải là số nguyên tố hay không?

```python
def is_prime_number(n) :
    if n < 2 :
        return False
    for i in range(2, int(n ** 0.5) + 1) :
        if n % i == 0 :
            return False
    return True
N = int(input())
print(is_prime_number(N))
```

### W5A4
VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên. 

```python
def is_perfect_number(n):
    t = 0 
    if n < 2 :
        return False 
    for i in range(1,n):
        if n % i == 0 :
            t += i
    if t == n :
        return True 
    else :
        return False
N = int(input())
print(is_perfect_number(N))
```

### W5A5
Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử 
đầu tiên có giá trị k trong list số nguyên, nếu không có thì trả về -1

```python
def pos_of_K(arr,k) :
    pos = -1
    vt = 0
    for x in arr :
        if x == k :
            return vt
        vt += 1
    return pos

assert pos_of_K([1,2,3,4,5,6,7], 8) == -1 
assert pos_of_K([1,2,3,4,5], 1) == 0
assert pos_of_K([1,2,3,4,5], 3) == 2
assert pos_of_K([1,2,3,4,5], 10) == -1

print("Pass all test")
```

### W5A6
Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!) 

```python
def factory(n) :
    gt = 1
    if n >= 0 :
        for i in range(1, n + 1):
            gt *= i 
        return gt 
    return "None"
assert factory(0) == 1
assert factory(3) == 6
assert factory(5) == 120
assert factory(1) == 1
assert factory(-10) == "None"

print("Pass all test")
```

Hàm thực hiện tính toán kết quả của phép toán giữa số \( num1 \) và \( num2 \) với toán tử \( operat 
\) 
Biết các toán tử đơn giản thực hiện phép toán lần lượt là: 
Phép trừ \( - \) 
Phép cộng \( + \) 
Phép chia \( / \) 
Phép nhân \( * \) 
Kết quả làm tròn đến chữ số thập phân thứ hai. 
Đầu vào: Hai số thực và phép toán cần thực hiện (trong số các phép toán \(+, -, *, /\)). 
Đầu ra: Kết quả phép toán. 

```python
def Calculator(a,b, operat) :
    if (operat == "+") :
        return a + b
    if (operat == "-") :
        return  a - b
    if (operat == "*") :
        return  a * b
    if operat == "/" :
        if b != 0 :
            x = a / b
            x = round(x, 2) 
            return x
    return "None"
a,op,b = input().split()
a = int(a)
b = int(b)
print(f"{a} {op} {b} = {Calculator(a,b,op)}")
#assert Calculator(8,9, "*") == 72
#assert Calculator(18,9, "/") == 2.00
```

### W5A8
Cho hai số nguyên  x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này. 
Khoảng cách Hamming là cái tên được đặt theo tên của Richard Hamming, người giới thiệu lý 
thuyết này trong tài liệu có tính cơ sở của ông về mã phát hiện lỗi và sửa lỗi (error-detecting and 
error-correcting codes). Nó được sử dụng trong kỹ thuật viễn thông để tính số lượng các bit trong 
một từ nhị phân (binary word) bị đổi ngược, như một hình thức để ước tính số lỗi xảy ra trong quá 
trình truyền thông, và vì thế, đôi khi, nó còn được gọi là khoảng cách tín hiệu (signal distance). Việc 
phân tích trọng số Hamming của các bit còn được sử dụng trong một số ngành, bao gồm lý thuyết 
tin học, lý thuyết mã hóa, và mật mã học. Tuy vậy, khi so sánh các dãy ký tự có chiều dài khác 
nhau, hay các dãy ký tự có xu hướng không chỉ bị thay thế không thôi, mà còn bị ảnh hưởng bởi dữ 
liệu bị lồng thêm vào, hoặc bị xóa đi, phương pháp đo lường phức tạp hơn, như khoảng cách 
Levenshtein (Levenshtein distance) là một phương pháp có tác dụng và thích hợp hơn. 

```python
def ten_to_bin(n) :
    bi = ""
    while n > 0 :
        bi = chr(n % 2 + ord("0")) + bi 
        n //= 2
    return bi 
n,m = map(int,input().split())
n_bin = ten_to_bin(n)
m_bin = ten_to_bin(m)
while len(n_bin) < len (m_bin) :
    n_bin = "0" + n_bin
while len(m_bin) < len (n_bin) :   
    m_bin = "0" + m_bin
cnt = 0
for i in range(len(n_bin)):
    if n_bin[i] != m_bin[i] :
        cnt += 1
print(n_bin, m_bin,cnt)

#assert ten_to_bin(1) == "1"
#assert ten_to_bin(4) == "100"
#assert ten_to_bin(6) == "110"
#assert ten_to_bin(10) == "1010"
#assert ten_to_bin(100) == "1100100"
#print("Pass all test")
#1 = 00000001
#4 = 00000100
```

### W5A9
Viết hàm trả về tổng các chữ số của số nguyên dương n

```python
def tongchuso(n) :
    tong = 0 
    while n > 0 :
        tong += n % 10 
        n //= 10
    return tong  
N = int(input())
print(tongchuso(N))
```

0 <= |r| < |divisor|.

Viết hàm compute(a, b, op) thực hiện phép tính a (+|-|*|/) b tùy theo ký tự op. 

```python
def compute(a,b,op):
    if op == "+" :
        return a + b
    if op == "-" :
        return a - b
    if op == "*" :
        return a * b
    if op == "/" :
        if b != 0 :
            x =  a /b
        return x 
a,b = map(int,input().split())
op = input()
print(compute(a,b,op))
```

### W5A15
Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông. 

```python
def is_right_triangle(a,b,c):
    if (a + b) > c and (a + c) > b and (b + c) > a :
        if a ** 2 == b ** 2 + c ** 2 or b ** 2 == a ** 2 + c ** 2 or c ** 2 == a ** 2 + b ** 2 :
            return True
        else :
            return False
a,b,c = map(int,input().split())
print(is_right_triangle(a,b,c))
```
