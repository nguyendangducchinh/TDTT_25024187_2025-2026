### Bai 4
Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím 

```python
h = int(input("Nhập độ cao h: "))
for i in range(1, h + 1):
    print(" " * (h - i), end="")
    if i == 1:
        print("*")
    elif i == h:
        print("*" * (2 * h - 1))
    else:
        print("*" + " " * (2 * i - 3) + "*")
```

### Bai 5
Nhập vào số nguyên dương n. Tính S = 1 + 2 + 3 + 4 + … + n

```python
n = int(input())
S = 0 
for i in range(1,n + 1):
    S += i
print(S)
```
