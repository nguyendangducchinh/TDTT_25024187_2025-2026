### Bai 4
Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím 

```python
h = int(input("Nhập độ cao h: "))
for i in range(1, h + 1):
    print(" " * (h - i), end="")
    if i == 1:
        print("*")
    elif i == h:
        print("*" * (2 * h - 1))
    else:
        print("*" + " " * (2 * i - 3) + "*")
```

### Bai 14
Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập 
xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng 


```python
dayso = map(int,input().split())
s = 0
max14 = 0
min14 = 0 
for n in dayso :
    if n == -1 :
        break
    s += 1
    if n > max14 :
        max14 = n
    if n < min14 :
        min14 = n
if s > 0:
    print(max14)
    print(min14)
else:
    print("Bạn chưa nhập số nào ngoài -1.") 
```

### Bài 15
 Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số. Làm bằng nhiều 
cách nhất có thể, ko dùng mảng. 

```python
n = int(input())
so_chu_so = 0
m = n
if m == 0 :
    so_chu_so = 1
else :
    while m > 0 :
        m //= 10
        so_chu_so += 1 
print(n,so_chu_so)
```

### Bai 16
Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu 
chữ số lẻ 

```python
n = int(input())
chan = 0
le = 0
if n == 0:
    chan = 1
else:
    m = n
    while m > 0:
        chu_so = m % 10
        if chu_so % 2 == 0:
            chan += 1
        else:
            le += 1
        m //= 10
print("Số chữ số chẵn:", chan)
print("Số chữ số lẻ:", le)
```

### Bai 18
Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 3^k hay 
không

```python
n = int(input())
if n <= 0:
    print("NO")
else:
    m = n
    while m % 3 == 0:
        m //= 3
    if m == 1:
        print("Yes")
    else:
        print("No")
```

### Bai 19
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số 
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong 
dãy số fibonacci lớn nhất nhưng không vượt quá A 


```python
A = int(input())
f1 = 1
f2 = 1
if A < 1:
    print("Không có số nào")
else:
    f_next = f1 + f2
    while f_next <= A:
        f1 = f2        
        f2 = f_next    
        f_next = f1 + f2 
    print("Số Fibonacci lớn nhất không vượt quá A là:", f2)
```

### Bai 20
Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố.

```python
n = int(input("Nhập số nguyên dương n: "))
dem = 0
if n == 0:
    dem = 0
else:
    m = n
    while m > 0:
        chu_so = m % 10  
        if chu_so == 2 or chu_so == 3 or chu_so == 5 or chu_so == 7:
            dem += 1
        m //= 10  
print(n,dem)
```
