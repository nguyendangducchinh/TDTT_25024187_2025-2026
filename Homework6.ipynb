### W6A19
Viết hàm đưa vào 1 list số nguyên, tìm và trả về vị trí có giá trị lớn nhất 
trong list 

```python
n = list(map(int,input().split()))
maxl = n[0]
vitri = 0
for i in n :
    if i > maxl :
        maxl = i
        vitri += 1
print(vitri)
```

### W6A20
Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và 
trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về

```python
n = list(map(int,input().split()))
k = int(input())
vt = 0
for i in range(len(n)) :
    if n[i] == k:
        vt += i
print(vt)
```

### W6A21
Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ 
chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện 

```python
n = map(str,input().split())
dct = {}
for i,x in enumerate(n) :
    if x not in dct :
        dct[x] = [i]
    else :
        dct[x].append(i)
for x in dct :
    dct[x] = tuple(dct[x])
print(dct)
```
