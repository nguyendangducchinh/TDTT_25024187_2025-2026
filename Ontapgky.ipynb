Yêu cầu: Cho tọa độ ba điểm A(x1,y1), B(x2,y2), C(x3,y3), trả về True nếu chúng tạo thành một tam giác, ngược lại trả về 
False. 
Quy tắc: Chúng tạo thành tam giác nếu không thẳng hàng (diện tích có hướng khác 0). 

```python
def kiem_tra_tam_giac(x1,y1,x2,y2,x3,y3):
    if (x2 - x1) * (y3 - y1) - (x3 - x1) * (y2 - y1) == 0 :
        return False
    else :
        return True
x1,y1,x2,y2,x3,y3 = map(int,input().split())
print(kiem_tra_tam_giac(x1,y1,x2,y2,x3,y3))
```

```python
def fibonacci_lon_nhat_khong_vuot_qua(a):
    if a <= 0:
        return "Vui lòng nhập số nguyên dương."
    f1, f2 = 1, 1
    while f2 <= a:
        f1, f2 = f2, f1 + f2
    return f1
n = int(input())
print(fibonacci_lon_nhat_khong_vuot_qua(n))
```

Yêu cầu: Cho cân nặng (kg) và chiều cao (m) của hai người, hãy trả về người có BMI cao hơn. 
Trả về 1 nếu người 1 có BMI cao hơn, 2 nếu người 2 có BMI cao hơn, và 0 nếu bằng nhau. 
Nhắc lại: BMI = weight / (height height). 

```python
def bmi(a1,b1,a2,b2) :
    bmi1 = a1 / (b1 * b1)
    bmi2 = a2 / (b2 * b2)
    if bmi1 > bmi2 :
        return 1
    elif bmi1 == bmi2 :
        return 0 
    else :
        return 2
m,n,p,q = map(float,input().split())
print(bmi(m,n,p,q))
```

### Bai 8 
Viết hàm trả về góc ∠ABC theo đơn vị độ khi biết tọa độ các điểm A(x1,y1), B(x2,y2), C(x3,y3). 

```python
import math
def timgoc(x1,y1,x2,y2,x3,y3) :
    a2 = (x2 - x1) ** 2 + (y2 - y1) ** 2
    b2 = (x3 - x2) ** 2 + (y3 - y2) ** 2
    c2 = (x1 - x3) ** 2 + (y1 - y3) ** 2
    a = a2 ** 0.5
    c = c2 ** 0.5
    cosA = (a2 + c2 - b2) / (2 * a * c)
    cosA = max(-1, min(1, cosA))
    A = math.acos(cosA)
    pi = 3.14159 
    gocA = A * (180 / pi)
    return gocA
x1,y1,x2,y2,x3,y3 = map(int,input().split())
print(timgoc(x1,y1,x2,y2,x3,y3))
```

- Gợi ý: Dùng phép kiểm tra diện tích (tích có hướng) và một sai số nhỏ khi so sánh số thực. Các điểm nằm trên cùng một

```python
def are_collinear(x1,y1,x2,y2,x3,y3):
    tichcohuong = ((x2 - x1) * (y3 - y1)) + ((x3 - x1) * (y2 - y1))
    return tichcohuong == 0 
x1,y1,x2,y2,x3,y3 = map(int,input().split())
print(are_collinear(x1,y1,x2,y2,x3,y3))
```

Yêu cầu: Viết hàm detect_number_type(s) trả về 'int' nếu s biểu diễn một số nguyên, 'float' nếu biểu diễn một số thực dấu 
chấm động, ngược lại trả về 'neither'. 
- Đầu vào: chuỗi s. 
- Đầu ra: một trong các giá trị {'int', 'float', 'neither'}. 
- Quy tắc: 
- Cho phép có dấu + hoặc - ở đầu. 
- Số nguyên phải chỉ gồm các chữ số và không có dấu chấm thập phân (ví dụ: '100'). 
- Số thực phải có dấu chấm và có chữ số ở cả hai phía của dấu chấm (ví dụ: '100.0', '-3.14'). 
- Các chuỗi như '.5', '5.', '1e3' được xem là 'neither' trong bài này. 

```python
def detect_number_type(s):
    if s == "" :
        return "neither"
    sobentrong = s
    if sobentrong[0] == "+" or sobentrong[0] == "-":
        sobentrong = sobentrong[1:]
    if sobentrong.isdigit():
        return "int"
    if "." in sobentrong :
        so = sobentrong.split(".")
        if len(so) == 2:
            if so[0].isdigit() and so[1].isdigit():
                return "float"
    return "neither"
s = input()
print(detect_number_type(s))
```

### Bai14
Yêu cầu: Viết hàm compute(a, b, op) tính a (+|-||/) b tùy theo op. 
- Đầu vào: các số a, b và ký tự op thuộc {'+', '-', '', '/'}. 
- Đầu ra: kết quả số học. 
- Yêu cầu thêm: - Nếu op là '/' và b == 0, phát sinh ValueError. 
- Nếu op không hợp lệ, phát sinh ValueError

```python
def compute(a,b,op) :
    if op == "+":
        return a + b
    if op == "-" :
        return a - b
    if op == "*" :
        return a * b
    if op == "/" :
        if b == 0 :
            raise ValueError
        return a / b 
```
