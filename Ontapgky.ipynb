### Bai 2
Yêu cầu: Cho số nguyên x, trả về True nếu tồn tại số nguyên n ≥ 1 sao cho x = 1 + 2 + ... + n, ngược lại trả về False. 
Khuyến khích dùng vòng lặp đơn giản để cộng dồn tổng. 

```python
def tong(x) :
    tong = 0 
    n = 1
    while tong < x  :
        tong += n
        n += 1
    if tong == x :
        return True
    else :
        return False
x = int(input())
print(tong(x))
```

### Bai 3
Yêu cầu: Cho tọa độ ba điểm A(x1,y1), B(x2,y2), C(x3,y3), trả về True nếu chúng tạo thành một tam giác, ngược lại trả về 
False. 
Quy tắc: Chúng tạo thành tam giác nếu không thẳng hàng (diện tích có hướng khác 0). 

```python
def kiem_tra_tam_giac(x1,y1,x2,y2,x3,y3):
    if (x2 - x1) * (y3 - y1) - (x3 - x1) * (y2 - y1) == 0 :
        return False
    else :
        return True
x1,y1,x2,y2,x3,y3 = map(int,input().split())
print(kiem_tra_tam_giac(x1,y1,x2,y2,x3,y3))
```

### Bai 4
Yêu cầu: Cho các hệ số thực a, b, c, trả về tất cả các nghiệm thực của phương trình bậc hai axx + bx + c = 0 trong một danh 
sách được sắp xếp tăng dần. 

```python
def nghiempt(a,b,c):
    if a == 0 :
        if b == 0 :
            if c == 0 :
                return "Phuong trinh vo so nghiem"
            else :
                return "Phuong trinh vo nghiem"
        else : 
            x = - c / b
            return x
    else :
        delta = b ** 2 - 4 * a * c
        if delta == 0 :
            x = - b / (2 * a)
            return [x]
        elif delta > 0 :
            x1 = (- b + delta ** 0.5 ) / (2 * a)
            x2 = (- b - delta ** 0.5 ) / (2 * a)
            if x1 < x2 :
                return [x1,x2]
            else :
                return [x2,x1]
        else :
            return "Phuong trinh vo nghiem"
a,b,c = map(int,input().split())
print(nghiempt(a,b,c))
```

### Bai 5
Yêu cầu: Cho ba số a, b, c, trả về chúng dưới dạng danh sách theo thứ tự không giảm. 

```python
def sapxep(a,b,c):
    if a > b :
        a,b = b,a
    if a > c :
        a,c = c,a
    if b > c : 
        b,c = c,b 
    return [a,b,c]
m,n,p = map(int,input().split())
print(sapxep(m,n,p))
```

Cho số nguyên ko âm,in ra số fibonanci thứ n

```python
def fibonacci_lon_nhat_khong_vuot_qua(a):
    if a <= 0:
        return "Vui lòng nhập số nguyên dương."
    f1, f2 = 1, 1
    while f2 <= a:
        f1, f2 = f2, f1 + f2
    return f1
n = int(input())
print(fibonacci_lon_nhat_khong_vuot_qua(n))
```

### Bai7
cho n là 1 số nguyên ko âm.In ra số fibonanci thứ n  
cho 1 số nguyên dương n.Kiểm tra xem n có phải là số fibonanci  
cho n >= 0 tính tổng n số fibonanci đầu tiên  
cho số nguyên dương a .Tìm số lớn nhất nhưng ko vượt quá a trong dãy fibonanci  

```python
#Bai1
def fibonanci1(n) :
    f0 = 0 
    f1 = 1
    while n > 0 :
        f2 = f0 + f1
        f0 = f1 
        f1 = f2 
        n -= 1
    return f0
n = int(input())
print(fibonanci1(n))
```

```python
#Bai2
def fibonanci2(n) :
    f0 = 0 
    f1 = 1
    while f0 < n :
        f2 = f0 + f1
        f0 = f1 
        f1 = f2 
    if f0 == n :
        return True
    else :
        return False
n = int(input())
print(fibonanci2(n))
```
