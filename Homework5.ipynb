### W5A16
Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số 
nguyên x 

```python
def kiemtraso(x) :
    k = len(str(x))
    tong = 0
    n = x
    while n > 0 :
        cs = n % 10 
        tong += cs ** k
        n //= 10
    if tong == x :
        return True
    else :
        return False
m = int(input())
print(kiemtraso(m))
```

Viết hàm kiêểm tra một số có phải số chính phương hay không? Hàm trả về kiểu bool, có tham số là một số 
nguyên x 

```python
def kiemtracp(x) :
    if x < 0 :
        return False
    for i in range(1, x + 1):
        if i * i == x :
            return True
    return False
n = int(input())
print(kiemtracp(n))
```

### W5A18
VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên.

```python
def doixung(n):
    m = n
    p = 0
    while m > 0 :
        cs = m % 10 
        p = p * 10 + cs
        m //= 10
    if p == n :
        return True
    else :
        return False
x = int(input())
print(doixung(x))
```

### W5A28
Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).

```python
import math
def goc(x1,y1,x2,y2,x3,y3) :
    ux, uy = x1 - x2 , y1 - y2
    vx, vy = x3 - x2 , y3 - y2
    dodai_u = math.isqrt(ux ** 2 + uy ** 2)
    dodai_v = math.isqrt(vx ** 2 + vy ** 2)
    uv = ux * vx + uy * vy
    cosA = uv / (dodai_u * dodai_v)
    cosA = max(-1, min(1, cosA))
    A = math.acos(cosA)
    gocA = math.degrees(A)
    return gocA
x1,y1,x2,y2,x3,y3 = map(int,input().split())
print(goc(x1,y1,x2,y2,x3,y3))
```

Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau 
Nếu n ≤ 0 trả về 0.

```python
def sum_1_to_n(n) :
    if n <= 0 :
        return 0 
    tong = 0 
    for i in range(1, n + 1) :
        chusodaco = []
        m = str(i)
        for chuso in m :
            if chuso not in chusodaco :
                chusodaco.append(chuso)
        if len(m) == len(chusodaco) :
            tong = tong + i
    return tong 
n = int(input())
print(sum_1_to_n(n))
```
