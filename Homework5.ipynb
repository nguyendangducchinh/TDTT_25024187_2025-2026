### W5A16
Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số 
nguyên x 

```python
def kiemtraso(x) :
    k = len(str(x))
    tong = 0
    n = x
    while n > 0 :
        cs = n % 10 
        tong += cs ** k
        n //= 10
    if tong == x :
        return True
    else :
        return False
m = int(input())
print(kiemtraso(m))
```

### W5A17
Viết hàm kiêểm tra một số có phải số chính phương hay không? Hàm trả về kiểu bool, có tham số là một số 
nguyên x 

```python
def kiemtracp(x) :
    if x < 0 :
        return False
    for i in range(1, x + 1):
        if i * i == x :
            return True
    return False
n = int(input())
print(kiemtracp(n))
```

### W5A18
VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên.

```python
def doixung(n):
    m = n
    p = 0
    while m > 0 :
        cs = m % 10 
        p = p * 10 + cs
        m //= 10
    if p == n :
        return True
    else :
        return False
x = int(input())
print(doixung(x))
```

Viết hàm is_one_product_of_others(a, b, c) trả về True nếu một trong ba số bằng tích của hai số còn lại 

```python
def is_one_product_of_others(a,b,c):
    if a * b == c or a * c == b or b * c == a :
        return True
    else:
        return False
a,b,c = map(int,input().split())
print(is_one_product_of_others(a,b,c))
```
