### W9A1
Viết hàm merge hai danh sách đã sắp xếp mà không dùng bất cứ thuật toán sắp xếp nào

```python
def merge_sorted_lists(list1, list2):
    i, j = 0, 0
    result = []
    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            result.append(list1[i])
            i += 1
        else:
            result.append(list2[j])
            j += 1
    while i < len(list1):
        result.append(list1[i])
        i += 1
    while j < len(list2):
        result.append(list2[j])
        j += 1
    return result

list1 = list(map(int, input().split()))
list2 = list(map(int, input().split()))
print(merge_sorted_lists(list1, list2))
```

### W9A2
Cho ma trận hai chiều nxm có các phần tử được sắp xếp trên dòng tăng dần và trên cột tăng dần, hãy tìm số lớn thứ k (k < nxm) với thuật toán tốt nhất có thể

```python
import heapq

def find_kth_largest_matrix(matrix, k):
    n = len(matrix)
    m = len(matrix[0])
    heap = []
    for r in range(n):
        for c in range(m):
            heap.append(-matrix[r][c])
    heapq.heapify(heap)
    
    res = 0
    for _ in range(k):
        res = heapq.heappop(heap)
    return -res

matrix = [[1, 5, 9], [10, 11, 13], [12, 13, 15]]
print(find_kth_largest_matrix(matrix, 8))
```

### W9A3
Cho một số nguyên dương, hãy thực hiện hoán đổi đúng 2 chữ số để được một số nguyên nhỏ nhất có thể

```python
def smallest_number_with_one_swap(num_str):
    num_list = list(num_str)
    n = len(num_list)
    
    for i in range(n):
        min_idx = i
        for j in range(n - 1, i, -1):
            if num_list[j] < num_list[min_idx]:
                if i == 0 and num_list[j] == '0':
                    continue
                if min_idx == i or num_list[j] < num_list[min_idx]:
                    min_idx = j
        if min_idx != i:
            num_list[i], num_list[min_idx] = num_list[min_idx], num_list[i]
            break
            
    return "".join(num_list)

s = input().strip()
print(smallest_number_with_one_swap(s))
```

- Xóa hết tất cả các chữ cái thường trong xâu a.

Cho 2 xâu a,b, xác đính xem có cách biến đổi từ xâu a sang xâu b không. Ghi ra "YES" nếu có và "NO" nếu ngược lại.

Đầu vào:
Dòng đầu tiên chứa 1 số nguyên Q là số lượng câu hỏi. Tương ứng với mỗi câu hỏi, đưa ra 2 xâu a,b trên 2 dòng khác nhau.(Q≤10,|a|,|b|≤1000)

Đầu ra:
Với mỗi câu hỏi, ghi ra kết quả tương ứng trên một dòng.

```python
def can_transform(a, b):
    n, m = len(a), len(b)
    dp = [[False] * (m + 1) for _ in range(n + 1)]
    dp[0][0] = True
    
    for i in range(n):
        for j in range(m + 1):
            if not dp[i][j]:
                continue
            if j < m and a[i].upper() == b[j]:
                dp[i+1][j+1] = True
            if not a[i].isupper():
                dp[i+1][j] = True
                
    return 'YES' if dp[n][m] else 'NO'

q = int(input())
for _ in range(q):
    a = input().strip()
    b = input().strip()
    print(can_transform(a, b))
```

Cho một dãy số gồm n chữ số là một hoán vị của dãy 1,2,...,n. Viết chương trình in ra hoán vị ở vị trí tiếp theo (danh sách các hoán vị được sắp xếp theo thứ tự từ điển) của hoán vị đã cho.

Đầu vào từ bàn phím gồm 2 dòng:
- Dòng đầu tiên chứa số nguyên n
- Dòng tiếp theo chứa n số nguyên là một hoán vị của dãy 1,2,...,n (n≤10^5)

Đầu ra: In ra màn hình hoán vị tiếp theo của hoán vị đã cho.

```python
def next_permutation(arr):
    n = len(arr)
    i = n - 2
    while i >= 0 and arr[i] >= arr[i+1]:
        i -= 1
    if i >= 0:
        j = n - 1
        while arr[j] <= arr[i]:
            j -= 1
        arr[i], arr[j] = arr[j], arr[i]
    
    left = i + 1
    right = n - 1
    while left < right:
        arr[left], arr[right] = arr[right], arr[left]
        left += 1
        right -= 1
    return arr

n = int(input())
arr = list(map(int, input().split()))
print(*(next_permutation(arr)))
```

Trước khi được giao cho bạn chăm sóc, trên luống hoa đã được trồng sẵn một số cây xanh tán rộng. Một ngày đầu xuân, công ty gửi đến cho bạn một số cây hoa và yêu cầu bạn trồng vào những vị trí còn trống trong luống hoa. Đây là những cây hoa hướng sáng và chúng không thể sống được dưới bóng râm của cây khác, đặc biệt là cây tán rộng. Và nếu trồng những cây hoa bên cạnh nhau, chúng sẽ tranh giành phân bón và nước và sau cùng, cả hai cây sẽ cùng chết.

Viết hàm canPlaceFlowers(flowerbed, k) kiểm tra xem có thể trồng hết số hoa mới được cấp vào trong luống đất của bạn hay không.

Hàm nhận đầu vào là một danh sách (list) flowerbed, đánh dấu những vị trí trong luống đất đã được trồng cây xanh trước đó (giá trị 1 ứng với vị trí cây đã được trồng và 0 ứng với vị trí đất trống) và số lượng hoa được phát thêm k (int). Hàm trả về True nếu có thể trồng hết toàn bộ k cây hoa vào luống đất và False trong trường hợp ngược lại.

```python
def canPlaceFlowers(flowerbed, k):
    count = 0
    n = len(flowerbed)
    for i in range(n):
        if flowerbed[i] == 0:
            prev_empty = (i == 0) or (flowerbed[i-1] == 0)
            next_empty = (i == n - 1) or (flowerbed[i+1] == 0)
            if prev_empty and next_empty:
                flowerbed[i] = 1
                count += 1
                if count >= k:
                    return True
    return count >= k
```

### W9A8
[Phát hiện đường đi lặp lại của Rô-bốt]
Ban đầu, một con rô-bốt ở vị trí (0, 0) trên một lưới 2D.
Cho một chuỗi `moves` chứa các thao tác di chuyển của rô-bốt.

Các bước di chuyển được biểu diễn bởi các ký tự:
- 'R' (Phải)
- 'L' (Trái)
- 'U' (Lên)
- 'D' (Xuống)

Viết hàm `def is_path_crossing(moves)`:
Hàm nhận đầu vào là một chuỗi `moves`.
Hàm trả về `True` nếu trong quá trình di chuyển, rô-bốt đi vào một ô mà nó đã từng đi qua trước đó.
Hàm trả về `False` nếu rô-bốt không bao giờ đi lặp lại (luôn đi vào ô mới).

Lưu ý: Vị trí ban đầu (0, 0) được tính là vị trí đầu tiên đã đi qua.

```python
def is_path_crossing(moves):
    x, y = 0, 0
    visited = {(0, 0)}
    for move in moves:
        if move == 'R': x += 1
        elif move == 'L': x -= 1
        elif move == 'U': y += 1
        elif move == 'D': y -= 1
        if (x, y) in visited:
            return True
        visited.add((x, y))
    return False
```

Viết hàm `def Median(nums):`
Hàm nhận đầu vào là một danh sách (list) `nums` được đảm bảo có đúng 5 số nguyên.

Yêu cầu:
Tìm và trả về số trung vị (số lớn thứ ba) của 5 số này.

Ràng buộc:
1. Chỉ được sử dụng tối đa 6 phép so sánh (ví dụ: <, >, <=, >=).
2. KHÔNG ĐƯỢC DÙNG bất kỳ hàm sắp xếp nào (như `sorted()`, `list.sort()`).

Chú ý: Bạn chỉ cần viết hàm `Median` theo yêu cầu.
Phần đọc input và gọi hàm (hàm main) đã có sẵn.

```python
def Median(nums):
    a, b, c, d, e = nums
    if a > b: a, b = b, a
    if c > d: c, d = d, c
    if a > c: a, c, b, d = c, a, d, b
    if e > c:
        if e > d: c, d, e = c, d, e
        else: d, e = e, d
    else:
        a, c, d, e = a, e, c, d
    if b > c:
        return d if b > d else b
    else:
        return c
```

### W9A10
[Tìm bán kính lò sưởi nhỏ nhất]

Mùa đông đang đến! Bạn được cung cấp vị trí của các ngôi nhà và các lò sưởi trên một đường thẳng.
Nhiệm vụ của bạn là tìm bán kính r nhỏ nhất cho các lò sưởi sao cho mọi ngôi nhà đều nằm trong phạm vi sưởi ấm của ít nhất một lò sưởi.

Tất cả các lò sưởi đều có cùng một bán kính r.
Một ngôi nhà tại vị trí h được sưởi ấm bởi một lò sưởi tại vị trí t nếu |h - t| <= r.

Viết hàm `def findRadius(houses, heaters):`
Hàm nhận đầu vào là hai danh sách (list) các số nguyên:
- `houses`: chứa vị trí của các ngôi nhà.
- `heaters`: chứa vị trí của các lò sưởi.
Hàm trả về một số nguyên là bán kính r nhỏ nhất cần thiết.

Lưu ý:
- Số lượng nhà và lò sưởi là một số không âm và không vượt quá 25000.
- Vị trí của các ngôi nhà và lò sưởi là một số không âm và không vượt quá 10^9.

```python
import bisect

def findRadius(houses, heaters):
    heaters.sort()
    max_radius = 0
    for house in houses:
        idx = bisect.bisect_left(heaters, house)
        dist1 = heaters[idx] - house if idx < len(heaters) else float('inf')
        dist2 = house - heaters[idx-1] if idx > 0 else float('inf')
        max_radius = max(max_radius, min(dist1, dist2))
    return max_radius
```

Yêu cầu:
- Không dùng vòng lặp
- Không dùng hàm có sẵn

```python
def power(base, exp):
    if exp == 0:
        return 1
    half = power(base, exp // 2)
    if exp % 2 == 0:
        return half * half
    else:
        return base * half * half
```

### W9A16
Cho hai số nguyên dương, $n$ và $m$ ($0 < n, m \le 20$). Bạn hãy in ra tất cả các chuỗi nhị phân có $n$ số 0 và $m$ số 1. In theo thứ tự tăng dần trong từ điển.

```python
def generate_binary_strings(n, m):
    result = []
    
    def backtrack(zeros, ones, current_str):
        if zeros == n and ones == m:
            result.append(current_str)
            return
        
        if zeros < n:
            backtrack(zeros + 1, ones, current_str + "0")
        if ones < m:
            backtrack(zeros, ones + 1, current_str + "1")
            
    backtrack(0, 0, "")
    for s in result:
        print(s)

n, m = map(int, input().split())
generate_binary_strings(n, m)
```

### W9A17
Viết hàm tìm dãy con tăng nghiêm ngặt dài nhất trong dãy đầu vào từ bàn phím.

```python
def longest_increasing_subsequence(arr):
    if not arr:
        return []
        
    n = len(arr)
    dp = [1] * n
    parent = [-1] * n
    
    for i in range(1, n):
        for j in range(0, i):
            if arr[i] > arr[j] and dp[i] < dp[j] + 1:
                dp[i] = dp[j] + 1
                parent[i] = j
                
    max_len = max(dp)
    max_idx = dp.index(max_len)
    
    lis = []
    while max_idx != -1:
        lis.append(arr[max_idx])
        max_idx = parent[max_idx]
        
    lis.reverse()
    return lis

arr = list(map(int, input().split()))
print(*(longest_increasing_subsequence(arr)))
```

### W9A21
Tìm phần tử lớn thứ $k$ trong dãy $n$ phần tử bằng thuật toán tốt nhất có thể – Tránh dùng sắp xếp (dùng phân hoạch của Quicksort).

```python
def partition(arr, low, high):
    pivot = arr[high]
    i = low
    for j in range(low, high):
        if arr[j] >= pivot:
            arr[i], arr[j] = arr[j], arr[i]
            i += 1
    arr[i], arr[high] = arr[high], arr[i]
    return i

def quick_select(arr, low, high, k):
    if low <= high:
        pivot_index = partition(arr, low, high)
        
        if pivot_index == k - 1:
            return arr[pivot_index]
        elif pivot_index > k - 1:
            return quick_select(arr, low, pivot_index - 1, k)
        else:
            return quick_select(arr, pivot_index + 1, high, k)
    return -1

def find_kth_largest(arr, k):
    return quick_select(arr, 0, len(arr) - 1, k)

arr = list(map(int, input().split()))
k = int(input())
print(find_kth_largest(arr, k))
```

### W9A22

```python
def is_anagram(s1, s2):
    if len(s1) != len(s2):
        return False
        
    freq1 = {}
    freq2 = {}
    
    for char in s1:
        freq1[char] = freq1.get(char, 0) + 1
    for char in s2:
        freq2[char] = freq2.get(char, 0) + 1
        
    return freq1 == freq2

s1 = input().strip()
s2 = input().strip()
if is_anagram(s1, s2):
    print("YES")
else:
    print("NO")
```

### W9A23
Cho số nguyên rất lớn (string), và một số nguyên dương $k$. Tìm số có $k$ chữ số lớn nhất có thể từ số ban đầu bằng cách xóa bớt đi một số chữ số nào đó mà không thay đổi vị trí của bất kì chữ số nào.

```python
def largest_number_after_deleting(num_str, k):
    to_delete = len(num_str) - k
    stack = []
    
    for digit in num_str:
        while to_delete > 0 and stack and stack[-1] < digit:
            stack.pop()
            to_delete -= 1
        stack.append(digit)
        
    final_str = ""
    for i in range(k):
        final_str += stack[i]
        
    return final_str

num_str = input().strip()
k = int(input())
print(largest_number_after_deleting(num_str, k))
```

Giả sử mật khẩu mạnh là một chuỗi kí tự dài ít nhất 8 kí tự gồm ít nhất 1 chữ cái thường, 1 chữ
cái hoa, 1 chữ số và 1 trong các kí tự đặc biệt !, @, #, $. Nhập vào một chuỗi kĩ tự, kiểm tra xem
nó có phải là mật khẩu mạnh không

```python
def is_strong_password(password):
    if len(password) < 8:
        return False
        
    has_lower = False
    has_upper = False
    has_digit = False
    has_special = False
    
    special_chars = "!@#$"
    
    for i in range(len(password)):
        char = password[i]
        
        if 'a' <= char <= 'z':
            has_lower = True
        elif 'A' <= char <= 'Z':
            has_upper = True
        elif '0' <= char <= '9':
            has_digit = True
        else:
            # Kiểm tra xem ký tự có thuộc danh sách ký tự đặc biệt yêu cầu không
            for j in range(len(special_chars)):
                if char == special_chars[j]:
                    has_special = True
                    break
                    
    return has_lower and has_upper and has_digit and has_special

input_password = input().strip()
print(is_strong_password(input_password))
```
