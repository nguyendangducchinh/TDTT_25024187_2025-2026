### Bai 3
Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là 
độ dài ba cạnh hay không, nếu  không thì in  ra màn  hình  'Khong la ba  canh 
cua mot tam  giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, 
Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn 
hình.  

```python
import math
a,b,c=map(float,input().split())
if a+b>c and a+c>b and b+c>a:
    p=(a+b+c)/2    
    print("Diện tích tam giác:", math.sqrt(p*(p-a)*(p-b)*(p-c)))
    print("Chu vi tam giác:",a+b+c)
    if a==b==c:
            print("Đây là tam giác đều")
    elif a==b or b==c or a==c:
            print("Đây là tam giác cân")
    elif a**2==b**2+c**2 or b**2==a**2+c**2 or c**2==a**2+b**2:
            print("Đây là tam giác vuông")
    else:
            print("Đây là tam giác thường")
else:
    print("Không phải là tam giác")
```

### Bai 5 
Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
a,b,c = map(float,input().split())
import math
if a == 0:
    if b == 0:
        if c == 0:
            print("Phương trình có vô số nghiệm.")
        else:
            print("Phương trình vô nghiệm.")
    else:
        x = -c / b
        print(f"Phương trình bậc nhất có nghiệm duy nhất: x = {x}")
else: 
    delta = b**2 - 4*a*c
    if delta > 0:
        x1 = (-b + math.sqrt(delta)) / (2*a)
        x2 = (-b - math.sqrt(delta)) / (2*a)
        print("Phương trình có 2 nghiệm phân biệt:")
        print(f"x1 = {x1}")
        print(f"x2 = {x2}")
    elif delta == 0:
        x = -b / (2*a)
        print(f"Phương trình có nghiệm kép: x1 = x2 = {x}")
    else:
        print("Phương trình vô nghiệm thực.")
```

### Bai 6
Tìm giá trị lớn nhất trong 4 số a, b, c, d (a, b, c, được nhập từ bàn phím).  

```python
a,b,c,d = map(int,input().split())
max = a 
if b > a :
    max = b
elif c > b :
    max = c 
elif d > c:
    max = d
print(max)
    
```
