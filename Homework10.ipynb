### W10A1
[DivideByZero]
Nhập vào hai số nguyên a, b. Thực hiện in ra kết quả phép chia a/b. Bắt lỗi ngoại lệ khi b bằng 0. Kết quả in ra chính xác tới 2 chữ số thập phân nếu b khác 0. (Dùng try... except bắt lỗi ZeroDivisionError)

```python
try:
    a = int(input())
    b = int(input())
    res = a / b
    print(f"{res:.2f}")
except ZeroDivisionError:
    print("ZeroDivisionError")
```

### W10A3
[ValueError - HappyNumbers]
Nhập vào một số nguyên dương a, Kiểm tra xem a có phải là số hạnh phúc không. Một số nguyên dương n được gọi là số hạnh phúc nếu ta thực hiện liên tiếp các bước sau: Tính tổng bình phương các chữ số của n. Lặp lại bước đó với kết quả vừa tính. Nếu cuối cùng kết quả là 1, thì n là số hạnh phúc. Nếu rơi vào vòng lặp vô hạn không bao giờ ra 1 (ví dụ như $4\rightarrow16\rightarrow37\rightarrow...$), thì n không phải là số hạnh phúc. Hãy nhập thử đầu vào là "a" rồi đọc kỹ Traceback để gỡ lỗi nhập liệu. (Dùng try... except bắt lỗi ValueError thêm set)

```python
try:
    n = int(input())
    vis = set()
    is_happy = False
    while n != 1 and n not in vis:
        vis.add(n)
        n = sum(int(c) ** 2 for c in str(n))
    if n == 1:
        is_happy = True
    print(is_happy)
except ValueError:
    print("ValueError")
```

### W10A4
Sử dụng try except, kết hợp cùng else... finally. Tính tổng 2 số, nếu nhập sai kiểu thì thông báo, luôn luôn in "Kết thúc chương trình." sau khi chạy.

```python
try:
    a = float(input())
    b = float(input())
except ValueError:
    print("Sai kieu")
else:
    print(a + b)
finally:
    print("Ket thuc chuong trinh.")
```

### W10A5
Viết chương trình chuyển tất cả phần tử của một list cho trước về int. Nếu phần tử nào trong list không thể chuyển được thì in cảnh báo và bỏ qua phần tử đó

Viết chương trình nhập một số x và tính căn bậc hai của nó. Bắt lỗi ngoại lệ khi x là số âm. Kết quả in ra chính xác tới 2 chữ số thập phân nếu x hợp lệ. (Dùng try... except bắt lỗi ValueError in ra "So am")

### W10A10
Viết chương trình nhập vào phần tử N ($N>0$) và một mảng gồm N số nguyên dương. Một mảng là hợp lệ nếu nó gồm các số khác nhau. In ra tổng các phần tử nếu mảng đó là hợp lệ. Nếu mảng không hợp lệ, in ra "Mang khong hop le".

### W10A14
[TypeError - SumNumbers]
Nhập vào một danh sách gồm nhiều giá trị. Hãy cộng tất cả phần tử sau khi chuyển được sang int. Nếu phần tử nào không chuyển được thì bỏ qua. In ra tổng cuối cùng.

```python
src = input().split()
total = 0
for x in src:
    try:
        total += int(x)
    except (ValueError, TypeError):
        continue
print(total)
```

### W10A15
[Raise_Exception - Positive Number]
Nhập vào một số nguyên n. Nếu $n<=0$ thì chủ động phát sinh ngoại lệ và in ra "So phai duong". Nếu hợp lệ thì in ra bình phương của n.

```python
try:
    n = int(input())
    if n <= 0:
        raise ValueError("So phai duong")
    print(n ** 2)
except ValueError as e:
    print("So phai duong")
```
